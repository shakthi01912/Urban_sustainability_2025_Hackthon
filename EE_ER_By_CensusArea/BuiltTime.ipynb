```python
import pandas as pd


df = pd.read_csv('/Users/shakthimaha/Downloads/Dataset/dwelling2014-lsoa.csv')
df_energy = pd.read_csv('/Users/shakthimaha/development/Urban/EE_ER_By_CensusArea/Revised_Energy_Efficiency_Cleaned.csv')


df_filtered = df[df["GEOG"] == "LSOA"]


valid_lsoa_values = set(df_energy['Lower_Super_Output_Area'])


df_final = df_filtered[df_filtered['lsoa'].isin(valid_lsoa_values)]


df_final.to_csv('/Users/shakthimaha/Downloads/Dataset/dwelling2014-lsoa_filtered.csv', index=False)

print("Final filtered dataset saved successfully!")


df_final.head(5)


```

```python
# Ensure merge columns have the same type and remove extra spaces
df_energy["Lower_Super_Output_Area"] = df_energy["Lower_Super_Output_Area"].astype(str).str.strip()
df_final["lsoa"] = df_final["lsoa"].astype(str).str.strip()

# Perform the merge
df_merged = df_energy.merge(df_final, left_on="Lower_Super_Output_Area", right_on="lsoa", how="inner")

# Select only the required columns
selected_columns = [
    'Revised_Energy_Efficiency_Rating',  
    'Pre_1900', '1900_1918', '1919_1929', '1930_1939', '1945_1954', 
    '1955_1964', '1965_1972', '1973_1982', '1983_1992', '1993_1999', '2000_2009', '2010_2014'
]

# Keep only the selected columns
df_merged = df_merged[selected_columns]

# Save the merged result to an Excel file
df_merged.to_csv("merged_data.csv", index=False)

print("Merge successful! File saved as 'merged_data.xlsx'.")
```
